```python
# While Loop Example
# print 0 to 10 numbers
temp = 0
while temp < 11:
    print(temp,end = ' ')
    temp +=1 # temp=temp+1    
```

```python
# Even numbers b/w 0 to 21 using while loop
t= 1
while t<21:
    if t%2==0:
        print(t,end = ' ')
    t+=1
```

```python
# print decrement numbers b/w 2 dynamic integers using while loop
a, b = int(input("Enter A Value= ")),int(input("Enter B Value= "))
if a > b:
    temp = a
    while temp > b:
        print(temp,end =' ')
        temp -=1
```

```python
# Infinate Loop without using increment/Decrement
temp = 0
while temp<10:
    print(temp,end = ' ')
```

```python
# Infinate loop using True
temp = 1
while True:
    print(temp,end= ' ')
    temp+=1
```

```python
# continue Example
# print odd numbers b/w 40 to 50
for temp in range(40, 51):
    if temp%2== 0:
        continue # Skip the iteration without terminate loop
    print(temp,end = ' ')    
        
```

```python
# break Example
# print 5 even numbers b/w 1 to 30 numbers
count = 0
for i in range(1,31):
    if i %2==0:
        count+=1
        print(i)
    if count==5:
        break # terminate the loop(for/while)
print("Out of loop")   
```

```python
a,b =int(input("Enter Lowest Value= ")),int(input("Enter Highest Value= "))
n = int(input("Enter N Value= "))
if a<b:
    count = 0;temp=0 
    while count<n:
        if temp%2==0:
            count+=1
            print(temp)
        if count==n:
            break
        temp+=1
```
